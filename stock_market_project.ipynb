```python
import requests
from bs4 import BeautifulSoup

def get_stock_data(ticker):
    url = f"https://finance.yahoo.com/quote/{ticker}/financials"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    # Find the data containing the fundamental
    data = soup.find("M(0) Whs(n) BdEnd Bdc($seperatorColor) D(itb)", {"class": "D(tbrg)"})

    # Extract the data from the table
    results = {}
    #          data.select may also work
    for row in data.find_all("tr"):
        key = row.find("td", {"class": "C($primaryColor) Fz(14px)"}).text
        value = row.find("td", {"class": "Ta(end) Fw(b) Lh(14px)"}).text
        results[key] = value

    return data

ticker = input("Enter the ticker symbol of the stock: ")
results = get_stock_data(ticker)
print(results)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_stock_fundamentals(ticker):
    url = f"https://finance.yahoo.com/quote/{ticker}/financials"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find("div", {"class": "D(tbrg)"})
    rows = table.find_all("div", {"class": "rw-expnded"})
    for row in rows:
        label = row.find("div", {"class": "C($primaryColor) Fz(s) Fw(500)"}).text
        value = row.find("div", {"class": "Ta(end) Fw(b) Lh(14px)"}).text
        print(f"{label}: {value}")

scrape_stock_fundamentals("AAPL")
```

```python
import requests
from bs4 import BeautifulSoup

def get_stock_data(ticker):
    url = f"https://stockanalysis.com/stocks/{ticker}/financials/"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    # Find the table containing the fundamental data
    table = soup.find("w-full whitespace-nowrap", {"class": "w-full whitespace-nowrap"})

    # Extract the data from the table
    data = {}
    for row in table.find_all("tr"):
        key = row.find("th", {"class": "C($primaryColor) Fz(14px)"}).text
        value = row.find("th", {"class": "Ta(end) Fw(b) Lh(14px)"}).text
        data[key] = value

    return data

ticker = input("Enter the ticker symbol of the stock: ")
data = get_stock_data(ticker)
print(data)
```
