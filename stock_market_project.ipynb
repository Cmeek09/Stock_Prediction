```python
import requests
from bs4 import BeautifulSoup

def get_stock_data(ticker):
    url = f"https://finance.yahoo.com/quote/{ticker}/financials"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    # Find the data containing the fundamental
    data = soup.find("M(0) Whs(n) BdEnd Bdc($seperatorColor) D(itb)", {"class": "D(tbrg)"})

    # Extract the data from the table
    results = {}
    #          data.select may also work
    for row in data.find_all("tr"):
        key = row.find("td", {"class": "C($primaryColor) Fz(14px)"}).text
        value = row.find("td", {"class": "Ta(end) Fw(b) Lh(14px)"}).text
        results[key] = value

    return data

ticker = input("Enter the ticker symbol of the stock: ")
results = get_stock_data(ticker)
print(results)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_stock_fundamentals(ticker):
    url = f"https://finance.yahoo.com/quote/{ticker}/financials"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find("div", {"class": "D(tbrg)"})
    rows = table.find_all("div", {"class": "rw-expnded"})
    for row in rows:
        label = row.find("div", {"class": "C($primaryColor) Fz(s) Fw(500)"}).text
        value = row.find("div", {"class": "Ta(end) Fw(b) Lh(14px)"}).text
        print(f"{label}: {value}")

scrape_stock_fundamentals("AAPL")
```

```python
import pandas as pd
import requests

api_key = 'K9OPGNS5IDET8VU5'
tickers = ['AAPL', 'MSFT', 'GOOGL', 'META', 'AMZN']  # list of tickers to retrieve income statement data for
income_statement_dict = {}  # create an empty dictionary to store income statement data for each ticker

for ticker in tickers:
    try:
        url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={api_key}'
        r = requests.get(url)
        data = r.json()
        df = pd.DataFrame(data['annualReports'])
        df = df.set_index('fiscalDateEnding').transpose()
        df = df.reset_index()
        income_statement_dict[ticker] = df
    except KeyError:
        print(f"No annual report data found for {ticker}. Skipping.")

income_statement_df = pd.concat(income_statement_dict)  # convert the dictionary to a DataFrame
income_statement_df.to_excel('income_statements.xlsx')
```

```python
import os.path
import pandas as pd
import requests

api_key = 'K9OPGNS5IDET8VU5'
tickers = ['DUK', 'SO', 'AEP', 'EXC', 'XEL']  # list of tickers to retrieve income statement data for

# Check if the file already exists
if os.path.isfile('income_statements10.xlsx'):
    # If it does, read the existing data into a DataFrame
    income_statement_df = pd.read_excel('income_statements.xlsx', index_col=0)
else:
    # Otherwise, create an empty DataFrame
    income_statement_df = pd.DataFrame()


# Loop over the tickers and retrieve the income statement data
for ticker in tickers:
    try:
        url = f'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={api_key}'
        r = requests.get(url)
        data = r.json()
        df = pd.DataFrame(data['annualReports'])
        df = df.set_index('fiscalDateEnding').transpose()
        df = df.reset_index()
        df.insert(0, 'Ticker', ticker)  # Add a column for the ticker symbol
        income_statement_df = income_statement_df.append(df, ignore_index=True)
    except KeyError:
        print(f"No annual report data found for {ticker}. Skipping.")

# Write the updated DataFrame to the file
income_statement_df.to_excel('income_statements10.xlsx', index=False)
```

```python
# Load the data into a pandas DataFrame
df = pd.read_excel('Data\IDX Stocks.xlsx')

# Reshape the DataFrame using melt
melted_df = pd.melt(df, id_vars=['symbol', 'type'], var_name='date', value_name='value')

# Pivot the table to get the desired format
pivoted_df = pd.pivot_table(melted_df, index=['symbol', 'date'], columns='type', values='value')

# Reset the index to get the date as a column
pivoted_df = pivoted_df.reset_index()

# Reorder the columns as desired
pivoted_df = pivoted_df[['symbol', 'Total Debt', 'Current Debt', 'Net Debt', 'Net Income', 'Total Revenue', 'Dividend Per Share', 'date']]


pivoted_df = pivoted_df.drop(columns=['Dividend Per Share'])

# Save the DataFrame to an Excel file
pivoted_df.to_excel('pivoted_data.xlsx', index=False)
```
